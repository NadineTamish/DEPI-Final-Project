 2. Data Exploration

```python
# (A) Analyze Image Resolutions
resolutions = []
for sample in dataset.take(100):  # Analyze first 100 images
    image = sample["image"].numpy()
    resolutions.append(image.shape[:2])  # (height, width)

# Convert to numpy array for analysis
resolutions = np.array(resolutions)
unique_resolutions = np.unique(resolutions, axis=0)

print("Unique image resolutions in dataset:", unique_resolutions)

#the dataset has minor variations in image sizes.
#If the model requires fixed sizes, we may need resizing or padding before training.
```

```python
target_size = (375, 1242)

def preprocess_image(image):
    return cv2.resize(image.numpy(), target_size)

resized_images = [preprocess_image(sample["image"]) for sample in dataset.take(10)]
```

```python
# Analyze Object Class Distribution
object_counter = Counter()
# Count object types in the dataset
for example in dataset.take(5000):  # Analyze 500 images
    labels = example["objects"]["type"].numpy()
    object_counter.update(labels)

# Print object type distribution
print("Object Class Distribution in KITTI Dataset:")
for class_id, count in object_counter.items():
    print(f"Class {class_id}: {count}")
```

```python
# KITTI class mapping
kitti_classes = {
    0: "Car",
    1: "Pedestrian",
    2: "Cyclist",
    3: "Van",
    4: "Truck",
    5: "Tram",
    6: "Misc",
    7: "DontCare"
}
```

```python
def plot_image_with_boxes(example):
    image = example["image"].numpy()
    bboxes = example["objects"]["bbox"].numpy()
    labels = example["objects"]["type"].numpy()  # Get object class IDs
    
    plt.figure(figsize=(8, 6))
    plt.imshow(image)

    # Convert normalized bbox to pixel coordinates
    height, width, _ = image.shape
    for bbox, label_id in zip(bboxes, labels):
        ymin, xmin, ymax, xmax = bbox
        xmin, xmax = xmin * width, xmax * width
        ymin, ymax = ymin * height, ymax * height

        # Draw bounding box
        plt.gca().add_patch(
            plt.Rectangle((xmin, ymin), xmax - xmin, ymax - ymin,
                        edgecolor='red', facecolor='none', linewidth=2)
        )

        # Add label text
        label_name = kitti_classes.get(label_id, "Unknown")
        plt.text(xmin, ymin - 5, label_name, color='red', fontsize=10,
                bbox=dict(facecolor='white', alpha=0.6, edgecolor='none'))

    plt.show()

# Load one sample image with labeled bounding boxes
sample = next(iter(dataset))
plot_image_with_boxes(sample)
```

```python
#check lighting, weather, and road types.visualize different images
for i, example in enumerate(dataset.take(5)):  # Show 5 random images
    print(f"Image {i+1}")
    plot_image_with_boxes(example)
```

```python
# Count occurrences of each object class
class_counts = {kitti_classes[label]: 0 for label in kitti_classes.keys()}

for example in dataset:
    labels = example["objects"]["type"].numpy()
    for label in labels:
        class_name = kitti_classes.get(label, "Unknown")
        class_counts[class_name] += 1

# Convert to DataFrame for visualization
df = pd.DataFrame(list(class_counts.items()), columns=["Class", "Count"])

# Plot the class distribution
plt.figure(figsize=(10, 5))
sns.barplot(x="Class", y="Count", data=df, palette="viridis")
plt.title("Class Distribution in KITTI Dataset")
plt.xticks(rotation=45)
plt.show()
```

```python
# Collect width and height of bounding boxes
widths = []
heights = []

for example in dataset.take(5000):  # Limit to 500 images for faster processing
    image = example["image"].numpy()
    h, w, _ = image.shape
    bboxes = example["objects"]["bbox"].numpy()
    
    for bbox in bboxes:
        ymin, xmin, ymax, xmax = bbox
        width = (xmax - xmin) * w
        height = (ymax - ymin) * h
        widths.append(width)
        heights.append(height)

# Convert to DataFrame
bbox_df = pd.DataFrame({"Width": widths, "Height": heights})

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.kdeplot(data=bbox_df, x="Width", y="Height", fill=True, cmap="magma")
plt.title("Heatmap of Bounding Box Sizes")
plt.show()
```

```python
#Bounding Box Size Distribution
bbox_sizes = []

for sample in dataset:
    image = sample["image"]
    height, width = image.shape[0], image.shape[1]
    
    for box in sample["objects"]["bbox"]:
        x_min, y_min, x_max, y_max = box.numpy() * [width, height, width, height]  # Convert to pixel coordinates
        bbox_width = x_max - x_min
        bbox_height = y_max - y_min
        bbox_sizes.append([bbox_width, bbox_height])

bbox_sizes = np.array(bbox_sizes)

# Plot heatmap of bounding box sizes
plt.figure(figsize=(8,6))
sns.kdeplot(x=bbox_sizes[:,0], y=bbox_sizes[:,1], cmap="magma", fill=True)
plt.xlabel("Width")
plt.ylabel("Height")
plt.title("Heatmap of Bounding Box Sizes")
plt.show()
```

```python
#Bounding Box Validity Check
invalid_boxes = 0
total_boxes = 0

for sample in dataset:
    image = sample["image"]
    height, width = image.shape[0], image.shape[1]
    
    for box in sample["objects"]["bbox"]:
        x_min, y_min, x_max, y_max = box.numpy() * [width, height, width, height]  # Convert to pixel coordinates
        total_boxes += 1
        if x_min < 0 or y_min < 0 or x_max > width or y_max > height:
            invalid_boxes += 1

print(f"Total bounding boxes: {total_boxes}")
print(f"Invalid bounding boxes: {invalid_boxes} ({(invalid_boxes/total_boxes)*100:.2f}%)")
```
