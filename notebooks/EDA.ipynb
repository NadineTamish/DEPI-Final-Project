```python
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt

# Load the dataset
dataset = tfds.load("kitti", split="train", data_dir="C:\\Users\\ramzy\\tensorflow_datasets")

# Take a single sample from the dataset
dataset = dataset.take(1)

for example in dataset:
    # Extract the image
    image = example["image"]
    
    # Extract objects (bounding boxes, dimensions, etc.)
    objects = example["objects"]

    # Display the image
    plt.imshow(image.numpy())
    plt.axis("off")
    plt.show()

    # Debug: Print the structure of `objects`
    print("Objects Structure:", type(objects))
    print(objects)

    # Iterate through detected objects
    for obj in objects:
        print("\nObject:", obj)

        # Handle different possible structures
        if isinstance(obj, dict) or hasattr(obj, 'keys'):
            print("Object bbox:", obj["bbox"].numpy())
            print("Object dimensions:", obj["dimensions"].numpy())
            print("Object location:", obj["location"].numpy())
        elif hasattr(obj, "bbox"):  # If `obj` is an object with attributes
            print("Object bbox:", obj.bbox.numpy())
            print("Object dimensions:", obj.dimensions.numpy())
            print("Object location:", obj.location.numpy())
        else:
            print("Unknown object structure:", obj)
```
